```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as snr
```

```python
data = pd.read_csv('/content/diabetes.csv')
```

```python
data.shape
```

```python
data.describe
```

```python
data.head()
```

```python
x = data.drop(['Outcome'], axis =1)
y = data['Outcome']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, train_size=.8)
```

```python
from sklearn.linear_model import LogisticRegression
log_model = LogisticRegression()
```

```python
log_model.fit(x_train, y_train)
```

```python
prediction = log_model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, prediction)
ac = accuracy_score(y_test, prediction)
```

```python
print(cm)
print(ac)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=7)
```

```python
knn.fit(x_train, y_train)
```

```python
k_pridiction = knn.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, k_pridiction)
ac = accuracy_score(y_test, k_pridiction)
```

```python
print(cm)
print(ac)
```

```python
from sklearn.tree import DecisionTreeClassifier
tree_model = DecisionTreeClassifier()
```

```python
tree_model.fit(x_train, y_train)
```

```python
t_prediction = tree_model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, t_prediction)
ac = accuracy_score(y_test, t_prediction)
```

```python
print(cm)
print(ac)
```

```python
from sklearn.ensemble import RandomForestClassifier
forest_model = RandomForestClassifier()
```

```python
forest_model.fit(x_train, y_train)
```

```python
f_prediction = forest_model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, f_prediction)
ac = accuracy_score(y_test, f_prediction)
```

```python
print(cm)
print(ac)
```

```python
from sklearn.naive_bayes import GaussianNB
bayes = GaussianNB()
```

```python
bayes.fit(x_train, y_train)
```

```python
b_prediction = bayes.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, b_prediction)
ac = accuracy_score(y_test, b_prediction)
```

```python
print(cm)
print(ac)
```

```python
from sklearn.svm import SVC
svc = SVC()
```

```python
svc.fit(x_train, y_train)
```

```python
s_prediction = svc.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, s_prediction)
ac = accuracy_score(y_test, s_prediction)
```

```python
print(cm)
print(ac)
```

```python
#using kernel
from sklearn import svm
svm = svm.SVC(kernel = 'linear')
```

```python
svm.fit(x_train, y_train)
```

```python
sv_prediction = svm.predict(x_test)
```
