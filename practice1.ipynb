#Exploratory Data Analysis

Gender

```python
 n_data.columns
```

```python
ax = snr.countplot(x='sex', data = n_data)
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
ax = snr.countplot(x = 'sex', data = n_data)
for bars in ax.containers:
  ax.bar_label(bars)
```

In this dataset count of male and female almost equal but compare to female male user count is slightly high.

```python
#using groupby to create graph.
n_data.groupby(['sex'], as_index = False)['expenses'].sum().sort_values(by = 'expenses', ascending = False)
ax = snr.barplot(x = 'sex', y = 'expenses', data = n_data)
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
n_data.columns
```

```python
bills = n_data.groupby(['region'], as_index = False)['expenses'].sum().sort_values(by = 'expenses', ascending = False)
ax = snr.barplot(x = 'region', y = 'expenses', data = bills)
for bars in  ax.containers:
  ax.bar_label(bars)
```

```python
bill = n_data.groupby(['sex'], as_index = False)['expenses'].sum().sort_values(by = 'expenses', ascending = False)
ax = snr.barplot(x = 'sex', y = 'expenses', data = bill)
for bars in ax.containers:
  ax.bar_label(bars)
```

In this graph male insurance expenses is more compare to female insurance expenses.

```python
import numpy as np
import pandas as pd
import seaborn as snr
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/content/tips.csv')
```

```python
data.info()
```

```python
data.describe()
```

```python
data.columns
```

```python
day = data.groupby(['day'], as_index = False)['total_bill'].sum().sort_values(by = 'total_bill', ascending = False)
ax = snr.barplot(x = 'day', y = 'total_bill', data = day)
snr.set(rc = {'figure.figsize' : (10,8)})
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
data.columns
```

```python
time = data.groupby(['time'], as_index = False)['tip'].sum().sort_values(by = 'tip', ascending = False)
tm = snr.barplot(x = 'time', y = 'tip', data = time)
snr.set(rc={'figure.figsize' : (12,8)})
for bars in tm.containers:
  tm.bar_label(bars)
```

```python
data.columns
```

```python
size = data.groupby(['size'], as_index = False)['total_bill'].sum().sort_values(by = 'total_bill', ascending = False)
sz = snr.barplot(x = 'size', y = 'total_bill' , data = size)
snr.set(rc ={'figure.figsize' : (15,7)})
for bars in sz.containers:
  sz.bar_label(bars)
```

```python
data.columns
```

```python
day = data.groupby(['sex'], as_index = False)['tip'].sum().sort_values(by = 'tip', ascending = False)
d = snr.barplot(x = 'sex', y = 'tip', data = day)
snr.set(rc={'figure.figsize' : (13,6)})
for bars in d.containers:
  d.bar_label(bars)
```
