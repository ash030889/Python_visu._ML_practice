```python
import numpy as np
import pandas as pd
import seaborn as snr
import matplotlib.pyplot as plt
```

```python
data = snr.load_dataset('iris')
```

```python
data.info()
```

```python
new_data = data.drop(['species'], axis = 1)
```

```python
new_data.head()
```

```python
ax = new_data.plot(figsize = (15,8), title = 'visualization of iris')
```

```python
ax.set_xlabel('axis = 1')
```

```python
ax.set_ylabel('axis = 11')
```
